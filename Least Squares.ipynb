```python
#Function fot find aprox by least squares
#Statistics Inference course
import numpy as np
import matplotlib.pyplot as plt
import math as mt

#datos de entrada
x = np.array([0. , 0.25, 0.5 , 0.75, 1. ])
y = np.array([1.95, 2.79, 3.51 , 4.21 , 5.03])
n=np.size(x)

#funcion para hallar w0 y w1
def regresion_lineal(x, y):
    n = np.size(x)
    w1 = (n*sum(x*y) - sum(x)*sum(y))/(n*sum(x**2)-sum(x)**2)
    w0 = (sum(y)-w1*sum(x))/n
    print(w0)
    print(w1)
    return w0, w1

a=regresion_lineal(x, y)

#funcion con parametros w0,w1
z = np.zeros(n)
for i in range(n):
    z[i]=a[0]+a[1]*x[i]
    
#funcion y=3x+2, funcion real para comparar
b = np.zeros(n)
for i in range(n):
    b[i]=3*x[i]+2


plt.plot(x, z, 'r')
plt.plot(x, b, 'b')
plt.legend(['z(x,w0,w1)', 'b(x)=2x+3'])
```
