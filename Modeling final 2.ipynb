# Modeling behavior while varying alpha parameter


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

def model1(t, v, alpha, beta, G, p, Lmax, lambd, delta, pc, k, K, mu, R, Ymax):
    L,S,C = v
    Y = alpha*beta*S*L+lambd*(1-beta)**delta*L**delta
    dLdt = (G*Y/L*(1-C)*(1-L/Lmax)-mu)*L
    dSdt = (p*(S/k-1)*(1-S/K)*(1-C)-alpha*beta*L/Lmax)*S
    dCdt = (R*Y/Ymax*(1-C)-pc*S)*C
    return (dLdt,dSdt,dCdt)
# Variables:
# S: Renewable recourses
# L: Labor force
# C: Contamination
# Parameters:
G= 5.025495331607045*10**(-5)  # G: Labor force growth
p= 1056.522565475319
alpha = 2453.384961728254  # alpha: Tech factor
beta=0.498813025194561 # beta: Preference factor
# K: Capital
k = 0.25 #min
K = 1    #max
lambd = 1.025486438597085*10**(-4)  # lambd: Proportionality factor
R = 0.2  # R: factor that increase the production
delta = 2.331111485964386  # delta: Elasticity coefficient
Lmax= 4932054.573601058 # Lmax: Max force that region can reach
pc = 0.014813725565046 # pc: regeneration rate of the enviorment
mu= 0.140268237915137  # mu: Mortality and/or migratory rate
#psi: Elasticity coefficient, we chose psi=1
psi=1
Ymax = alpha*beta*Lmax+lambd*(1-beta)**delta*Lmax**delta

t_in=0
t_fin=100
t = np.linspace(t_in, t_fin, 1000)
sol = solve_ivp(model1, [t_in, t_fin], [556000, 0.50222284902714, 0.02],args=[alpha, beta, G, p, Lmax, lambd, delta, pc, k, K, mu, R, Ymax], t_eval=t,dense_output=True)

v = sol.sol(t)

fig, ax = plt.subplots()
ax.plot(t,v[2])
plt.title("Contamination rate vs Time")
plt.xlabel("Time")
plt.ylabel("C")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t,v[1])
plt.title("Renewable recourses vs Time")
plt.xlabel("Time")
plt.ylabel("S")
plt.show()
```
