# Modelado con parametro de tabla 2


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp

def model1(t, v, alpha, beta, G, p, Lmax, lambd, delta, pc, k, K, mu, R, Ymax):
    L,S,C = v
    Y = alpha*beta*S*L+lambd*(1-beta)**delta*L**delta
    dLdt = (G*Y/L*(1-C)*(1-L/Lmax)-mu)*L
    dSdt = (p*(S/k-1)*(1-S/K)*(1-C)-alpha*beta*L/Lmax)*S
    dCdt = (R*Y/Ymax*(1-C)-pc*S)*C
    return (dLdt,dSdt,dCdt)
# Variables:
# S: Recursos renovables
# L: Fuerza laboral
# C: Contaminacion
# Parametros:
G= 5.025495331607045*10**(-5)  # G: Tasa de crecimiento de la fuerza laboral
p= 1056.522565475319
alpha = 2453.384961728254# alpha: Factor tecnologico
beta=0.498813025194561 # beta: Factor de preferencia
k = 0.25 # K: Capital
K = 1
lambd = 1.025486438597085*10**(-4)  # lambd: factor de proporcionalidad
R = 0.2  # R: factor que incrementa ctte por produccion
delta = 2.331111485964386  # delta: coeficiente de elasticidad
Lmax= 4932054.573601058 # Lmax: Fuerza laboral maxima que puede soportar la region
pc = 0.014813725565046  # pc: tasa de regeneracion del medio
mu= 0.140268237915137 # mu: tasa de mortalidad o migracional
# psi: Coeficiente de elsaticidad, Por comodidad escogemos psi=1
Ymax = alpha*beta*Lmax+lambd*(1-beta)**delta*Lmax**delta

t_in=0
t_fin=100
t = np.linspace(t_in, t_fin, 1000)
sol = solve_ivp(model1, [t_in, t_fin], [556000, 0.50222284902714, 0.02],args=[alpha, beta, G, p, Lmax, lambd, delta, pc, k, K, mu, R, Ymax], t_eval=t,dense_output=True)

v = sol.sol(t)

fig, ax = plt.subplots()
ax.plot(t,v[2])
plt.title("Tasa de contaminacion")
plt.xlabel("Tiempo")
plt.ylabel("C")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(t,v[1])
plt.title("Tasa de recursos renovables")
plt.xlabel("Tiempo")
plt.ylabel("S")
plt.show()
```
